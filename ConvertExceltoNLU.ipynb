```python
import pandas as pd
import os

pd.set_option('max_colwidth', None)

PATH = os.getcwd()
```

```python
df = pd.read_excel(PATH + "/Data_ChatHDSD.xlsx", sheet_name="Sheet1")
# df.head(5)
```

```python
import html
# df["RESPONSE"] = df["RESPONSE"].apply(html.unescape)
df['RESPONSE'] = df['RESPONSE'].apply(lambda x: html.unescape(x) if isinstance(x, str) else x)
df['EXAMPLE'] = df['EXAMPLE'].apply(lambda x: html.unescape(x) if isinstance(x, str) else x)
```

```python
df['RESPONSE'] = df['RESPONSE'].str.replace("\"", "'")
df['RESPONSE'] = df['RESPONSE'].str.replace("*", "+")
```

```python
# Write Domain Response YML:
f1 = open(PATH + "/domain_yml_respon.txt", "w")

for i, utter in enumerate(df["STT"]):
    if not df["STT"].isnull()[i]:
        line = "  utter_" + str(int(utter)) + ":\n" + "    - text: " + "\"" + df["RESPONSE"][i] + "\"" + "\n" + "\n"
        f1.write(line)
        
    # else:
    #     line = "    - text: \"" + str(df["RESPONSE"][i]) + "\"\n"
    #     f1.write(line)

f1.close()
```

```python
# Write NLU YML:
f1 = open(PATH + "/nlu_yml.txt", "w")
f2 = open(PATH  + "/domain_yml_intent.txt", "w")
f2.write("intents:\n")

for i, intent in enumerate(df["STT"]):
    if not df["STT"].isnull()[i]:
        # print(intent)
        line = "\n- intent: " + "intent_" + str(int(intent)) + "\n" + "  examples: |\n" + "    - " + str(df["EXAMPLE"][i]) + "\n"
        f1.write(line)
        line = "- " + "intent_" + str(int(intent)) + "\n"
        f2.write(line)
        
    else:
        line = "    - " + str(df["EXAMPLE"][i]) + "\n"
        f1.write(line)

f1.close()
f2.close()
```
